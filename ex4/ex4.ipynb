### Partie 4 : Eulerian and Hamiltonian path

Dans cette partie nous allons étudier les grpahes eulerian et hamiltoniens.


#### Exercice 1 : Explication des problèmes

1. **Eulerian path**  
    Un chemin eulerian est un chemin d'un graphe qui passe une seule fois par chaque arrête du graphe. Concrètement si on peut tracer le graphe sans lever le crayon et sans passer 2 fois par le même trait tout en revenant au point de départ alors on est dans un chemin eulérien.  
    L'existence d'un chemin eulerian peut être déterminer à l'avance :  
    - Dans le graphe, il doit être possible d'accéder à chaque sommet depuis n'importe quel autre sommet. On dit que le graphe est connexe.  
    - Il doit y avoir 0 ou 2 sommets qui possèdent un nombre impair d'arrêtes qui les touchent.  
2. **Hamiltonian path**  
    Un chemin hamiltonian est un chemin qui passe une seule fois par chaque point d'un graphe. Il n'y a pas de conditions simples pour déterminer la présence d'un chemin hamiltonian dans un graphe. On est néanmoins capable de déterminer la présence de ce chemin dans certains graphes mais pas pour tous (ex: théorème de Dirac ou théorème d'Ore)  

3. **Graphe linéaire**
    Un graphe linéaire d'un graphe est un graphe dont chaque sommet correspond à une arrête du graphe initial. Par ailleurs 2 sommets du graphe linéaire sont adjacents si et seulement si leurs arrêtes correspondantes dans le graphe initial partagent un sommet commun.

4. **Lien entre chemin eulerian et hamiltonian**
    - Un chemin eulerian d'un graphe correspond à un chemin hamiltonian dans le graphe linéaire du graphe initial.  
    - Inversement si nous trouvons un chemin hamiltonian dans un graphe linéaire alors ce chemin correspond à un chemin eulerian dans le graphe initial.

#### Exercice 3: Quel algorithme est plus difficile ?  

D'un point de vue algorithmique le chemin eulerian est plus simple de par le fait que l'existence d'un chemin eulerian dans un graphe est facilement vérifiable. De plus l'algorithme pour trouver un chemin eulerian est un algorithme linéaire ce qui atteste de la simplicité en terme algorithmique.  
Le chemin Hamiltonian quant à lui est plus difficile à démontrer car il ne possède pas de condition prédéfini d'existence.  
Il est possible de vérifier en temps polynomial si une solution proposée est correcte ce qui fait du chemin hamiltonian un algorithme dans un "NP".
De plus il est possible de faire une réduction polynomial d'un problème connu NP-complet vers le chemin Hamiltonian (Ex: Hamiltonian Cycle). Ainsi celà fait du problème du chemin hamiltonian un problème NP-difficile. Etant un dans un NP et également un problème NP-difficile, le chemin Hamiltonian est donc un problème **NP-Complet** ce qui le rend bien plus complexe algorithmiquement parlant que le chemin eulerian.
